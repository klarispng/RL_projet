```python
#!/usr/bin/env python3
"""
Implmentation dun agent Double DQN avec action masking pour lenvironnement discrete de Highway.
La configuration est charge depuis un fichier, et toutes les valeurs des hyperparamtres sont directement
utilises dans le code (sans bloc ddi de dfinition).
"""

import gymnasium as gym
import pickle
import torch
import torch.nn as nn
import torch.nn.functional as F
import random
import numpy as np
from collections import deque
import matplotlib.pyplot as plt
import highway_env

# Chargement de la configuration
with open("config/1-highway-discrete-config.pkl", "rb") as f:
    config = pickle.load(f)

env = gym.make("highway-fast-v0", render_mode="human")
env.unwrapped.configure(config)
observation, _ = env.reset()

# --------------------------------------------------------------------------
# Fonctions d'aide
# --------------------------------------------------------------------------
def is_risky(state):
    """
    Dtermine si l'tat (observation) prsente un risque pour l'acclration.
    L'observation est issue d'une grille d'occupation.
    Utilise la configuration sous config["observation"]:
      - grid_size et grid_step se trouvent dans config["observation"]["grid_size"]
        et config["observation"]["grid_step"].
    On considre risqu si, dans la colonne la plus  droite de la grille, une cellule
    indique la prsence d'un vhicule (valeur > 0.5).
    """
    grid_range = config["observation"]["grid_size"][0]
    grid_step = config["observation"]["grid_step"][0]
    nx = int((grid_range[1] - grid_range[0]) / grid_step)
    ny = int((config["observation"]["grid_size"][1][1] - config["observation"]["grid_size"][1][0]) / config["observation"]["grid_step"][1])
    
    num_features = 7  # ['presence', 'x', 'y', 'vx', 'vy', 'cos_h', 'sin_h']
    
    if state.size != nx * ny * num_features:
        return False
    grid = state.reshape(nx, ny, num_features)
    front_presence = grid[-1, :, 0]
    if np.any(front_presence > 0.5):
        return True
    return False

def shape_reward(reward, action, state):
    """
    Applique le reward shaping :
      - Si la reward indique une collision (en comparant  config["collision_reward"]),
        on renforce la pnalit.
      - Si l'action est d'acclrer (action == 1) dans un contexte risqu, on applique un malus.
      - Un offset positif de +0.5 est ajout pour quilibrer le signal.
    """
    shaped = reward
    if reward <= config.get("collision_reward", -1):
        shaped = reward * 2
    if action == 1 and is_risky(state):
        shaped -= 0.5
    return shaped

# --------------------------------------------------------------------------
# Dfinition du rseau de neurones (DQN)
# --------------------------------------------------------------------------
class DQN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, output_dim)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)

# --------------------------------------------------------------------------
# Dfinition de l'agent (Double DQN avec action masking)
# --------------------------------------------------------------------------
class DQNAgent:
    def __init__(self, state_dim, action_dim):
        # Instanciation des rseaux
        self.q_net = DQN(state_dim, action_dim)
        self.target_net = DQN(state_dim, action_dim)
        self.target_net.load_state_dict(self.q_net.state_dict())

        # Optimiseur et scheduler (lr fix en dur)
        self.optimizer = torch.optim.Adam(self.q_net.parameters(), lr=1e-3)
        self.scheduler = torch.optim.lr_scheduler.StepLR(self.optimizer, step_size=100, gamma=0.9)

        # Replay buffer avec capacit directement intgre
        self.replay_buffer = deque(maxlen=100000)

        # Hyperparamtres dfinis "en dur" dans le code
        self.gamma = 0.99
        self.epsilon = 1.0
        self.epsilon_min = 0.05
        self.epsilon_decay = 0.995
        self.batch_size = 64
        self.action_dim = action_dim

    def get_action(self, state):
        # En exploration, on applique un action masking sur l'acclration
        if np.random.rand() < self.epsilon:
            available_actions = list(range(self.action_dim))
            if is_risky(state) and 1 in available_actions:
                available_actions.remove(1)
                if len(available_actions) == 0:
                    available_actions = [1]
            return random.choice(available_actions)
        # En exploitation, modification des Q-values pour viter l'acclration en contexte risqu
        state_tensor = torch.FloatTensor(state).unsqueeze(0)
        with torch.no_grad():
            q_values = self.q_net(state_tensor).squeeze(0).numpy()
        if is_risky(state):
            q_values[1] = -float('inf')
        return int(np.argmax(q_values))
    
    def update_epsilon(self, current_episode, total_episodes):
        # Aprs la moiti des pisodes, epsilon est fix  sa valeur minimale
        if current_episode > total_episodes // 2:
            self.epsilon = self.epsilon_min
        else:
            self.epsilon = max(self.epsilon * self.epsilon_decay, self.epsilon_min)

    def update(self):
        if len(self.replay_buffer) < self.batch_size:
            return

        batch = random.sample(self.replay_buffer, self.batch_size)
        states, actions, rewards, next_states, dones = map(np.array, zip(*batch))

        states = torch.FloatTensor(states)
        actions = torch.LongTensor(actions).unsqueeze(1)
        rewards = torch.FloatTensor(rewards).unsqueeze(1)
        next_states = torch.FloatTensor(next_states)
        dones = torch.FloatTensor(dones).unsqueeze(1)

        q_values = self.q_net(states).gather(1, actions)

        with torch.no_grad():
            # Double DQN : slection et valuation spares
            next_actions = self.q_net(next_states).argmax(dim=1, keepdim=True)
            next_q_values = self.target_net(next_states).gather(1, next_actions)
            target = rewards + self.gamma * (1 - dones) * next_q_values

        loss = F.mse_loss(q_values, target)
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

# --------------------------------------------------------------------------
# Fonction de training
# --------------------------------------------------------------------------
def train(agent, env):
    all_rewards = []
    num_episodes = 500      # Nombre d'pisodes hard-cod
    sync_target_every = 10 # Frquence de synchronisation du rseau cible

    for ep in range(num_episodes):
        state, _ = env.reset()
        state = state.flatten()  # Ajustement de la forme
        total_reward = 0
        done = False

        while not done:
            action = agent.get_action(state)
            next_state, reward, terminated, truncated, _ = env.step(action)
            next_state = next_state.flatten()
            done = terminated or truncated

            shaped_reward = shape_reward(reward, action, state)
            agent.replay_buffer.append((state, action, shaped_reward, next_state, done))
            agent.update()

            state = next_state
            total_reward += reward  # On cumule la reward d'origine

        agent.update_epsilon(ep, num_episodes)
        agent.scheduler.step()
        all_rewards.append(total_reward)

        if ep % sync_target_every == 0:
            agent.target_net.load_state_dict(agent.q_net.state_dict())

        current_lr = agent.optimizer.param_groups[0]['lr']
        if ep % 10 == 0:
            print(f"Episode {ep}: Reward = {total_reward:.2f}, Epsilon = {agent.epsilon:.4f}, LR = {current_lr:.6f}")

    return all_rewards

def plot_rewards(rewards):
    plt.plot(rewards)
    plt.xlabel("Episode")
    plt.ylabel("Total Reward")
    plt.title("Training Performance")
    plt.show()

# --------------------------------------------------------------------------
# Partie principale
# --------------------------------------------------------------------------
if __name__ == "__main__":
    # Dtermination de la dimension de l'tat et du nombre d'actions
    state_dim = observation.flatten().shape[0]
    action_dim = env.action_space.n

    agent = DQNAgent(state_dim, action_dim)
    rewards = train(agent, env)
    plot_rewards(rewards)
```
